```python
import matplotlib.pyplot as plt 
import numpy as np 
MB_CSTN = 0.00000095367432
encryption_times = np.array([3438248,6857826,10155451,14183503,17746205,20847037,24114317,27458981,32384693,36185237])/(10**6)
decryption_times = np.array([2890966,5670732,8255940,11734012,14546532,16854262,19118821,22069321,24991355,27564017])/(10**6)
sizes = [1600000 ,3200000 ,4800000 ,6400000, 8000000 ,9600000 ,11200000 ,12800000, 14400000 ,16000000]
sizes_mb = np.array(sizes)*MB_CSTN
```

```python
plt.plot(sizes_mb,encryption_times)
plt.plot(sizes_mb, decryption_times)
plt.ylabel('Time[ms]')
plt.xlabel('Size [MB]')
plt.title("Encrypion and decryption times in [ms] \n 100 iterations average")
plt.savefig("enc_dec_bench.jpg")
plt.show()
```
